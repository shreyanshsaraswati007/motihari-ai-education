## Combine Datasets for East Champaran District

### Subtask:
Merge the `df_census`, `df_internet`, and `df_udise` DataFrames into a single cohesive dataset for the East Champaran district.

**Reasoning**:
I will aggregate the `df_udise` DataFrame to calculate district-level metrics for East Champaran, summing the specified columns and averaging the pupil-teacher ratio, as per the instructions.



```python
df_udise_aggregated = pd.DataFrame({
    'District': ['East Champaran'],
    'Total_Schools_District': [df_udise['Total_Schools'].sum()],
    'Enrollment_Primary_District': [df_udise['Enrollment_Primary'].sum()],
    'Enrollment_Secondary_District': [df_udise['Enrollment_Secondary'].sum()],
    'Teachers_Primary_District': [df_udise['Teachers_Primary'].sum()],
    'Teachers_Secondary_District': [df_udise['Teachers_Secondary'].sum()],
    'Pupil_Teacher_Ratio_District_Avg': [df_udise['Pupil_Teacher_Ratio'].mean()]
})

print("--- Aggregated UDISE Data for East Champaran (df_udise_aggregated) Head ---")
print(df_udise_aggregated.head())
print("\n--- Aggregated UDISE Data Info ---")
df_udise_aggregated.info()
```

**Reasoning**:
Now, I will merge the `df_east_champaran_district` DataFrame with the `df_udise_aggregated` DataFrame on the 'District' column to create the final combined dataset for East Champaran, as per the instructions. Finally, I will display its head and info to verify its structure and content.



```python
df_east_champaran_combined = pd.merge(df_east_champaran_district, df_udise_aggregated, on='District', how='inner')

print("--- Final Combined East Champaran Data (df_east_champaran_combined) Head ---")
print(df_east_champaran_combined.head())
print("\n--- Final Combined East Champaran Data Info ---")
df_east_champaran_combined.info()
```

### Caveat: Normalization of a Single Row

When `MinMaxScaler` is applied to a DataFrame containing only a single row (as is the case with `df_east_champaran_combined` after filtering for East Champaran district), the output for all numerical features will be `0.0`.

This occurs because `MinMaxScaler` scales values according to the formula: `(value - min) / (max - min)`. In a single-row dataset, the `min` and `max` for each feature are the same as the `value` itself. Consequently, the numerator `(value - min)` becomes `0`, and the denominator `(max - min)` also becomes `0`. The `MinMaxScaler` implementation handles this by returning `0.0` for such cases.

While this satisfies the technical requirement of applying normalization, it's important to understand that in a real-world multi-row dataset, normalization is applied to scale values relative to the range observed across *all* samples, not just one. For a single row, the utility of normalization for comparative analysis is limited, as all features become `0.0` and lose their relative magnitudes to each other within that single row context. If `df_east_champaran_combined` had contained multiple districts, the normalization would have produced meaningful relative scales for each district's metrics.

## AI Modeling - Regression for Influencing Factors

### Subtask:
Implement a regression model to analyze how education outcomes (e.g., literacy rates) relate to other socioeconomic variables.


**Reasoning**:

```python
import pandas as pd

# Define the target variable (dependent variable)
y = df_district_combined['Literacy_Rate_Total']

# Define the feature variables (independent variables)
X = df_district_combined[['Internet_Penetration_Rate', 'Population', 'Gender_Ratio']]

print("--- Target Variable (y) Head ---")
print(y.head())
print("\n--- Feature Variables (X) Head ---")
print(X.head())
```

    *   This negative coefficient implies that for every one-unit increase in the 'Gender_Ratio' (meaning more females per 1000 males), the 'Literacy_Rate_Total' is expected to decrease by approximately 0.98 percentage points, holding other variables constant. This result might seem counter-intuitive at first glance. However, in the context of districts with lower female literacy historically, a higher gender ratio (more females) in certain regions might coincide with areas that have faced greater educational disparity for women in the past, thus potentially lowering the overall average literacy rate if those disparities are still prominent.
